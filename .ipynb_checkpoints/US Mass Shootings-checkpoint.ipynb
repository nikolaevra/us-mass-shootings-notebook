# US Mass Shootings #


## Part 2 - Importing the dataset and analysing data

```python
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Mass Shootings Dataset.csv')

dataset.iloc[115:120, :].style
```

## Part 3 - Starting to Analyze data

So lets start with some basic analytics:

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

m_illness = dataset.iloc[:, 7].values

# categorize data to make comparisons faster
le = LabelEncoder()
m_illness = le.fit_transform(m_illness)

# 4 categories that are present in the dataset
# Yes
cat_yes = 0
# No
cat_no = 0
# Unknown
cat_unknown = 0
# Unclear
cat_unclear = 0

for item in m_illness:
    if item == 0:
        cat_no += 1
    elif item == 2:
        cat_unknown += 1
    elif item == 3:
        cat_yes += 1
    elif item == 1:
        cat_unclear += 1

# create labels and colors
labels = 'Yes', 'No', 'Unknown', 'Unclear'
x = [cat_yes, cat_no, cat_unknown, cat_unclear]

# plot the results
plt.pie(x, labels=labels, autopct='%.2f')
plt.show()
```

Judge this graph for your self, but from what I see there is no distict dominance of people with mental illnesses that have caused mass shootings, so lets move on and look for some other attributes that might help us make the classification

```python
dead = dataset.iloc[:, 4].values
injured = dataset.iloc[:, 5].values

import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
area = np.pi * (15 * np.random.rand(N))**2

plt.scatter(dead[1:], injured[1:], s=area, c='blue', alpha=0.2)
plt.xlabel("Deaths")
plt.ylabel("Injuries")
plt.show()
```

**Note:** this scatter plot does not include one of the most recent shootings, the Las Vegas one

```python
plt.scatter(dead[1:], injured[1:], s=area, c='blue', alpha=0.2)
plt.scatter(dead[0], injured[0], s=area, c='red', alpha=1)
plt.annotate("Las Vegas (Oct 1st, 2017)", xy=(dead[0], injured[0]), xytext=(10, -50),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='grey', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
plt.xlabel("Deaths")
plt.ylabel("Injuries")
plt.show()
```
